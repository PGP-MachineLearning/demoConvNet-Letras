# Programa auxiliar para separar las imágenes para Entrenamiento y Prueba ('Train' & 'Test'):
El dataset de imágenes original es obtenido de http://www.ee.surrey.ac.uk/CVSSP/demos/chars74k/

1) Cargar librerías:

```python
import os
import os.path
import numpy as np
import shutil

print ("Librerías cargadas.")
```

3) Levantar y procesar la carpeta de imágenes a distribuir:

```python
# variables auxiliares
all_images_array = []
auxiPath = path + "/TodasLetras"
all_dirs = os.listdir(auxiPath)

# crea los directorios, si es necesario
if not os.path.isdir(path +'/Letras'):
  os.makedirs(path +'/Letras')
if not os.path.isdir(path +'/Letras/train'):
  os.makedirs(path +'/Letras/train')
if not os.path.isdir(path +'/Letras/test'):
  os.makedirs(path +'/Letras/test')

# procesa cada directorio
for each_dir in all_dirs:
  print("\n-- Procesando ", each_dir)

  allFileNames = []
  train_FileNames = []
  test_FileNames = []

  dirDestTrain = path +'/Letras/train/' + each_dir
  dirDestTest = path +'/Letras/test/' + each_dir

  # levanta la lista de imágenes de cada letra
  allFileNames = os.listdir("".join([auxiPath, "/", each_dir]))

  # desordena y separa 80% para train y 20% para test
  np.random.shuffle(allFileNames)
  train_FileNames, test_FileNames = np.split(np.array(allFileNames),
                                                          [int(len(allFileNames)*0.8)])

  print('Total: ', len(allFileNames))
  print('Training: ', len(train_FileNames))
  print('Testing: ', len(test_FileNames))

  # crea los directorios de entrenamiento y test
  if not os.path.isdir(dirDestTrain):
    os.makedirs(dirDestTrain)
  if not os.path.isdir(dirDestTest):
    os.makedirs(dirDestTest)

  # mueve a entrenamiento
  cant = 0
  for name in train_FileNames:
    if os.path.isfile(dirDestTrain + '/' + name):
      print("**", name, " ya existe en ", dirDestTrain)      
    else:
      shutil.move(auxiPath + '/'+ each_dir + '/' + name, dirDestTrain)
      cant = cant + 1
  print(cant, " imágenes movidas a ", dirDestTrain)

  # mueve a test
  cant = 0
  for name in test_FileNames:
    if os.path.isfile(dirDestTest + '/' +  name):
      print("**", name, " ya existe en ", dirDestTest)
    else:      
      shutil.move(auxiPath + '/'+ each_dir + '/' + name, dirDestTest)
      cant = cant + 1
  print(cant, " imágenes movidas a ", dirDestTest)

print("\n*** Distribución Terminada ***\n")
```
